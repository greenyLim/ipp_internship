### 파일 이름 변경 코드 : os 라이브러리 사용
파일 이름 변경 코드 - 앞에 번호가 6씩 증가해서 저장되어 있어서, 6씩 빼서 저장함

```python
# 파일 이름 변경 코드 - 앞에 번호가 6씩 증가해서 저장되어 있어서, 6씩 빼서 저장함


import os

# 파일이 있는 디렉토리 경로
directory = 'C:\현대백화점'

# 디렉토리 내의 파일 목록 가져오기
files = os.listdir(directory)

for filename in files:
    prefix, rest = filename.split('__', 1)
    print(prefix)
    print(rest)
    number = int(prefix)
    print(number)
    new_number = number - 6
    print(new_number)
    new_filename = f"{new_number:02d}__{rest}"
    
    # 파일 이름 변경
    os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
```

'__첨부파일1' 삭제 

```python
# '__첨부파일1' 삭제 

# 파일이 있는 디렉토리 경로
directory = 'C:\현대백화점'

# 디렉토리 내의 파일 목록 가져오기
files = os.listdir(directory)

for filename in files:
    # '__첨부파일1'로 끝나는 경우에만 변경
        new_filename = filename.replace('__첨부파일1', '')  # '__첨부파일1' 문자열을 제거하여 새로운 파일 이름 생성
        
        # 파일 이름 변경
        os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
        
        print(f"{filename} 파일이 {new_filename}으로 성공적으로 변경되었습니다.")
```

### 첨부파일 누락 번호 찾기 위해 파일들의 번호만 저장해서 lst로 저장
### 그 후에 lst에서 1-216번 중에 없는 번호를 추출 

```python
# 파일이 있는 디렉토리 경로
directory = 'C:\현대백화점'

# 디렉토리 내의 파일 목록 가져오기
files = os.listdir(directory)

lst=[]
for filename in files:
    prefix, rest = filename.split('__', 1)
    print(prefix)
    lst.append(int(prefix))
```

```python
lst.sort()

# 1부터 216까지의 숫자 중에서 리스트에 없는 숫자 찾기
missing_numbers = []
current_number = 1

for num in lst:
    while current_number < num:
        missing_numbers.append(current_number)
        current_number += 1
    current_number += 1

while current_number <= 216:
    missing_numbers.append(current_number)
    current_number += 1

print("리스트에 없는 숫자:", missing_numbers)
```

### 대표자명과 담당자명이 같은 경우, 중복처리하고 비고에 저장 
- 비고에 '{중복인 번호}와 중복'으로 저장

```python
import pandas as pd

# 중복된 항목 찾기
duplicates = df[df.duplicated(subset=['담당자명', '대표자명'], keep=False)]

# 중복된 항목의 번호를 저장할 딕셔너리 생성
duplicate_indices = {}

# 중복된 항목의 번호 저장
for idx, row in duplicates.iterrows():
    key = row['담당자명'] + "_" + row['대표자명']
    if key not in duplicate_indices:
        duplicate_indices[key] = []
    duplicate_indices[key].append(row['번호'])

# '비고' 열에 중복 여부와 중복된 항목의 번호 기록
df['비고'] = ''
for key, indices in duplicate_indices.items():
    for idx, row in df.iterrows():
        if row['담당자명'] + "_" + row['대표자명'] == key:
            df.at[idx, '비고'] = f"{', '.join(map(str, indices))}번과 중복"

print(df)
```
