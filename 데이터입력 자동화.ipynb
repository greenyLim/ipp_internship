### 노가다 복붙 1000개 데이터 입력 자동화 코드
- txt 파일로 우선 전체 복붙한다음, ',' 이나 enter키를 기준으로 분리하여 데이터 개별 요소로 저장.

```python
## enter키로만 분리인 경우

# 파일 경로 설정
file_path = "운영사.txt"

# 전체 기관 리스트 초기화
organizations = []

# 파일을 읽어서 기관을 리스트에 추가
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        # 줄의 양 끝에 있는 공백과 개행 문자 제거 후 추가
        organizations.append(line.strip())

# 전체 리스트 출력
print(organizations)
```

```python
org_list=organizations
```

```python
lst1 = [org for org in org_list if org]
lst1
```

```python
## ',' 와 enter키 함께 처리해야할 때

# 파일 경로 설정
file_path = "협력기관.txt"

# 전체 기관 리스트 초기화
lst = []

# 파일을 읽어서 기관을 리스트에 추가
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        # 쉼표로 구분하여 각 요소를 추가
        organizations = line.strip().split(',')
        # 리스트에 추가
        lst.extend(organizations)

# 전체 리스트 출력
print(lst)
```

```python
lst2 = [org for org in lst if org]
lst2
```

```python
import pandas as pd

# lst 자체를 한 열에 저장할 때

# lst1, lst2 데이터프레임으로 변환
df1 = pd.DataFrame({'Column': lst1})
df2 = pd.DataFrame({'Column': lst2})

# 두 데이터프레임을 concat 함수를 사용하여 결합
result_df = pd.concat([df1, df2], ignore_index=True)

# 결과 출력
print(result_df)
```
